```python
#gym_duckietown is very chatty
import logging
gym_duckietown.logger.setLevel(logging.WARNING)
```

```python
env = Simulator("straight_road",
    full_transparency=True,
    domain_rand=False,
    user_tile_start=[1,0],
    seed=42,
    max_steps=float("inf"),
    robot_speed=1.2,
)
```

```python
env.unwrapped.start_pose = [[0.7, 0.0, 0.2], 0]
```

```python
plt.imshow(env.reset())
```

```python
env.reset()
for _ in range(100):
    obs, reward, done, info = env.step([1,1])
    if reward == -1000:
        break
    if done:
        break
    print(info["Simulator"]["robot_speed"])
```
